```
datapath = 'https://raw.githubusercontent.com/Build-Spotify-flow/Data-science/master/Data/Dataset.csv'
df = pd.read_csv(datapath)
df.head()
```

```
features = ['danceability', 'energy', 'instrumentalness', 'speechiness', 'valence']
plotter = df[features]
```

```
plotter.describe()
```

```
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

song_index = 35
with plt.style.context('default'):
  fig = plt.figure()
  categories=list(plotter)
  N = len(categories)
  values=plotter.loc[song_index].values.flatten().tolist()
  values += values[:1]
  angles = [n / float(N) * 2 * pi for n in range(N)]
  angles += angles[:1]
  ax = plt.subplot(111, polar=True)
  plt.xticks(angles[:-1], categories, color='black', size=10)
  plt.ylim(0,1)
  ax.plot(angles, values, linewidth=1, linestyle='solid', c='black')
  ax.fill(angles, values, '#1DB954', alpha=0.5)
  # ax.set_title(df['artist_name'].iloc[song_index] + '\n' + df['track_name'].iloc[song_index])
  ax.set_facecolor('xkcd:grey')
```

```
import base64
import io 
pic_IObytes = io.BytesIO()
fig.savefig(pic_IObytes,  format='png')
pic_IObytes.seek(0)
pic_hash = base64.b64encode(pic_IObytes.read())
```
